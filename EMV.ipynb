## Estimador Máximo Verosímil

1. Una v.a $X$, es exponencial con parámetro $\lambda >0$ si
$$
\mathbb{P}(X) = 1 - e^{-\lambda x}, \quad x > 0
$$
Calcular el **EMV**:

La función de densidad de probabilidad de una distribución exponencial es:
$$
f(x; \lambda) = \lambda e^{-\lambda x}, \quad x > 0
$$

Si tenemos una muestra de tamaño $n$, $x_1, x_2, \dots, x_n$, la función de verosimilitud es:
$$
L(\lambda) = \prod_{i=1}^n \lambda e^{-\lambda x_i} = \lambda^n e^{-\lambda \sum_{i=1}^n x_i}
$$

El logaritmo de la verosimilitud es:
$$
\log L(\lambda) = n \log \lambda - \lambda \sum_{i=1}^n x_i
$$

Para encontrar el **EMV** de $\lambda$, derivamos con respecto a $\lambda$ e igualamos a cero:
$$
\frac{d}{d\lambda} \log L(\lambda) = \frac{n}{\lambda} - \sum_{i=1}^n x_i = 0
$$

Despejando $\lambda$:
$$
\hat{\lambda} = \frac{n}{\sum_{i=1}^n x_i}
$$

2. Supongamos que la vida útil de los focos de cierta marca se modela mediante una distribución exponencial con parámetro desconocido $\lambda$. Probamos 5 focos y encontramos que la vida útil de $2, 3, 1, 3$ y $4$ años respectivamente. ¿Cuál es el **EMV** para $\lambda$?

Usamos la fórmula del EMV:
$$
\hat{\lambda} = \frac{n}{\sum_{i=1}^n x_i} = \frac{5}{2 + 3 + 1 + 3 + 4} = \frac{5}{13} \approx 0.3846
$$

Por lo tanto, el EMV de $\lambda$ es $\hat{\lambda} \approx 0.3846$.

3. Suponga que se tiene una variable aleatoria discreta $X$ con densidad:

$$
P(X=0) = \frac{2}{3}\theta, \quad P(X=1) = \frac{1}{3}\theta, \quad P(X=2) = \frac{2}{3}(1-\theta), \quad P(X=3) = \frac{1}{3}(1-\theta)
$$

con $\theta \in [0,1]$. Se toman las siguientes $10$ observaciones independientes de acuerdo a la distribución dada:
$$
(3, 0, 2, 1, 3, 2, 1, 0, 2, 1).
$$
Encuentra el **EMV** de $\theta$:

La función de verosimilitud es:
$$
L(\theta) = \left( \frac{2}{3}\theta \right)^2 \left( \frac{1}{3}\theta \right)^3 \left( \frac{2}{3}(1-\theta) \right)^3 \left( \frac{1}{3}(1-\theta) \right)^2
$$

El logaritmo de la verosimilitud es:
$$
\log L(\theta) = 2 \log\left( \frac{2}{3}\theta \right) + 3 \log\left( \frac{1}{3}\theta \right) + 3 \log\left( \frac{2}{3}(1-\theta) \right) + 2 \log\left( \frac{1}{3}(1-\theta) \right)
$$

Derivando con respecto a $\theta$ e igualando a cero, resolvemos para encontrar el **EMV** de $\theta$.

## Intervalos de confianza

1. En una encuesta de la UP, 225 estudiantes son seleccionados al azar de los que toman cálculo y se les pregunta si cálculo es su materia favorita. 100 estudiantes responden que el cálculo es su materia favorita. Proporcione un intervalo de confianza del 95 % para la proporción de todos los estudiantes que toman cálculo y que la consideran su materia favorita.

La proporción muestral es:
$$
\hat{p} = \frac{100}{225} \approx 0.4444
$$

El intervalo de confianza del 95 % para una proporción se calcula como:
$$
\hat{p} \pm Z_{\alpha/2} \sqrt{\frac{\hat{p}(1 - \hat{p})}{n}}
$$

Donde $Z_{\alpha/2} = 1.96$ para un nivel de confianza del 95 %.

Calculamos el margen de error:
$$
ME = 1.96 \sqrt{\frac{0.4444(1 - 0.4444)}{225}} \approx 0.0648
$$

El intervalo de confianza es:
$$
0.4444 \pm 0.0648
$$

2. Los datos
$$
1, 2, 3, 4
$$
se toman a partir de una distribución normal $N(\mu,\sigma^2)$ con $\mu$ desconocida.

- **1. Encuentra un intervalo de confianza al nivel de $90\%$ para $\mu$, dado que $\sigma =2$:**

El intervalo de confianza para la media con varianza conocida es:
$$
\bar{x} \pm Z_{\alpha/2} \frac{\sigma}{\sqrt{n}}
$$

Donde $\bar{x} = \frac{1+2+3+4}{4} = 2.5$, $Z_{\alpha/2} = 1.645$ para un nivel de confianza del 90 %, $\sigma = 2$ y $n=4$.

Calculamos el margen de error:
$$
ME = 1.645 \times \frac{2}{\sqrt{4}} = 1.645
$$

El intervalo de confianza es:
$$
2.5 \pm 1.645 \Rightarrow (0.855, 4.145)
$$

- **2. Encuentra un intervalo de confianza al nivel de $90\%$ para $\mu$ sin conocer $\sigma$:**

Cuando $\sigma$ es desconocido, usamos la distribución $t$ de Student con $n-1$ grados de libertad. El intervalo de confianza es:
$$
\bar{x} \pm t_{\alpha/2, n-1} \frac{s}{\sqrt{n}}
$$

Donde $s$ es la desviación estándar muestral:
$$
s = \sqrt{\frac{(1-2.5)^2 + (2-2.5)^2 + (3-2.5)^2 + (4-2.5)^2}{3}} = 1.29
$$

El valor crítico $t_{0.05, 3} = 2.353$. Entonces el margen de error es:
$$
ME = 2.353 \times \frac{1.29}{\sqrt{4}} = 1.517
$$

El intervalo de confianza es:
$$
2.5 \pm 1.517 \Rightarrow (0.983, 4.017)
$$

- **3. Encuentra un intervalo de confianza al nivel de $90\%$ para $\sigma^2$:**

El intervalo de confianza para la varianza se basa en la distribución $\chi^2$. El intervalo de confianza es:
$$
\left( \frac{(n-1)s^2}{\chi^2_{\alpha/2, n-1}}, \frac{(n-1)s^2}{\chi^2_{1-\alpha/2, n-1}} \right)
$$

Con $n=4$, $s^2 = 1.67$, $\alpha = 0.10$, y los valores de $\chi^2$ son $\chi^2_{0.05, 3} = 7.815$ y $\chi^2_{0.95, 3} = 0.352$.

El intervalo de confianza es:
$$
\left( \frac{3 \times 1.67}{7.815}, \frac{3 \times 1.67}{0.352} \right) \Rightarrow (0.641, 14.23)
$$
